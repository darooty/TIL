서포트 벡터 머신은 선형 결정경계를 갖는 로지스틱 회귀 같은 퍼셉트론의 단점을 보완하여 안정적인 결정경계를 생성하는 머신러닝 기법입니다. 안정적인 결정경계를 이해하기 위해선 마진을 알아야 합니다. 결정경계는 다음과 같습니다.  

결정경계함수의 출력값 y를 +1, -1로 갖는 x를 각각 x+, x-라고 해봅시다. 나머지 x들은 1보다 크거나 -1보다 작은 값을 갖습니다. 결정경계와 x+, x-의 거리는 다음과 같습니다.  

두 거리의 합을 마진이라고 하며 마진값이 클수록 결정경계가 안정적으로 형성되었다고 판단합니다. 따라서 마진의 합을 최대로 하는 경우는 다음의 값을 최소로 하는 것과 같습니다.  

모든 x에 대해 결정경계 출력값이 1이상 또는 -1이하이므로 다음이 성립합니다.  

위 식을 라그랑주 승수법을 사용하면 손실함수를 얻을 수 있습니다. 이 식을 최소화하는 문제를 해결하면 최적의 w를 계산할 수 있습니다.


SVM은 커널의 종류를 선택하는 kernel과 슬랙변수 허용치를 결정하는 C를 필요로 합니다. C가 작을수록 슬랙변수의 허용량이 늘어나 일반화 성능이 향상됩니다. 로지스틱 회귀모형이나 서포트 벡터 머신의 문제는 비선형 문제를 해결할 수 없다는 것입니다. 서포트 벡터 머신은 커널 트릭을 사용하여 비선형 문제를 해결합니다. 많이 사용되는 커널은 RBF, Polynomial 커널, sigmoid 커널 등이 있습니다. 커널 svm을 사용하면 하이퍼 파라미터 gamma를 필요로 합니다. 이 값이 작을수록 개별 포인트의 영향력 범위가 커져서 일반화 성능이 향상됩니다. 자세한 하이퍼 파라미터는 다음을 참조하면 알 수 있습니다.  
Reference: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

사용할 예제 데이터는 캐글 컴패티션: 타이타닉의 데이터셋을 직접 가공한 데이터입니다. 클랜징 - 파생변수 생성(합성, 바이닝, 파싱 등) - 연속형 변수의 로그변환 - 순서형 변수의 라벨인코딩 - 등간변수의 원핫인코딩을 적용하였습니다. 여기서 주의할 점은 로지스틱 회귀분석의 가정은 단지, y가 이항분포(or 베르누이)를 따르며, 모수가 x에 의존한다는 것 뿐이라는 것입니다. 따라서 독립변수 간의 독립성이나 정규성이 만족되지 않아도 충분히 사용할 수 있습니다. 연속형 변수의 로그변환은 독립변수의 정규성 확보를 위한 태크닉이며 선형 회귀 모형에서 성능 향상을 도모할 수 있으나 이번 파트에서는 큰 영향이 없습니다.

```python
import numpy as np
import pandas as pd

X_train = pd.read_csv('../Data/titanic_xtrain.csv')
X_test = pd.read_csv('../Data/titanic_xtest.csv')
y_train = pd.read_csv('../Data/titanic_ytrain.csv')
y_test = pd.read_csv('../Data/titanic_ytest.csv')
```

모든 전처리 과정을 마친 훈련데이터입니다. 첫 컬럼은 사용자 id로 제거 대상입니다.

```python
X_train
```

```python
X_train.drop(['Unnamed: 0'], axis=1, inplace=True)
X_test.drop(['Unnamed: 0'], axis=1, inplace=True)
y_train = y_train.Survived
y_test = y_test.Survived
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, GridSearchCV, cross_val_score
from sklearn.metrics import roc_auc_score
```

```python
cv = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
```

```python
params_svc = {
    'C': np.logspace(-3, 2, 6),
    'gamma': np.logspace(-3, 2, 6),
    'kernel': ['rbf', 'poly', 'sigmoid']
}

search_svc = GridSearchCV(SVC(),params_svc, cv=cv, refit=True, verbose=2).fit(X_train ,y_train)
```

```python
cv_rlt = pd.DataFrame(search_svc.cv_results_)
cv_rlt.iloc[cv_rlt.rank_test_score.sort_values().index].head(5)
```

```python
print(roc_auc_score(y_test, SVC(C=100, gamma=0.01, kernel='rbf').fit(X_train, y_train).predict(X_test)))
print(roc_auc_score(y_test, SVC(C=10, gamma=0.01, kernel='rbf').fit(X_train, y_train).predict(X_test)))
print(roc_auc_score(y_test, SVC(C=1, gamma=0.1, kernel='poly').fit(X_train, y_train).predict(X_test)))
print(roc_auc_score(y_test, SVC(C=0.001, gamma=1, kernel='poly').fit(X_train, y_train).predict(X_test)))
print(roc_auc_score(y_test, SVC(C=1, gamma=0.1, kernel='rbf').fit(X_train, y_train).predict(X_test)))
```
